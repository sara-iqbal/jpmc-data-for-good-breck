```python
# Step 4: Data cleaning functions
def clean_dataframe(df):
    """Main data cleaning function"""

    # Create a copy
    df_clean = df.copy()

    print("Step 1: Removing metadata rows...")
    # Remove rows with "Updated" in first column
    if df_clean.shape[1] > 0:
        first_col = df_clean.columns[0]
        df_clean = df_clean[~df_clean[first_col].astype(str).str.contains('Updated', na=False)]

    print("Step 2: Cleaning column names...")
    # Clean column names
    new_columns = {}
    for i, col in enumerate(df_clean.columns):
        if pd.isna(col) or str(col).strip() == '':
            new_columns[col] = f'column_{i}'
        else:
            clean_name = str(col).strip()
            clean_name = re.sub(r'[^\w\s]', ' ', clean_name)
            clean_name = re.sub(r'\s+', '_', clean_name)
            clean_name = clean_name.lower().strip('_')
            new_columns[col] = clean_name

    df_clean = df_clean.rename(columns=new_columns)

    print("Step 3: Removing empty rows and columns...")
    # Remove completely empty columns and rows
    df_clean = df_clean.dropna(axis=1, how='all')
    df_clean = df_clean.dropna(how='all')

    print("Step 4: Cleaning numeric columns...")
    # Clean numeric columns (1-5 scale questions)
    numeric_patterns = [
        'understanding.*assembly',
        'impactful.*grooming',
        'seeing.*words',
        'discussing.*safety',
        'reflecting.*age',
        'wanting.*share',
        'confident.*knowledge'
    ]

    for col in df_clean.columns:
        for pattern in numeric_patterns:
            if re.search(pattern, col, re.IGNORECASE):
                df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
                # Ensure values are between 1-5
                df_clean[col] = df_clean[col].apply(lambda x: x if 1 <= x <= 5 else np.nan)
                print(f"   - Cleaned: {col}")

    print("Step 5: Cleaning text columns...")
    # Clean text columns
    text_columns = [col for col in df_clean.columns if any(word in col for word in ['learn', 'aspect', 'explain', 'know_more', 'add'])]

    for col in text_columns:
        df_clean[col] = df_clean[col].astype(str)
        df_clean[col] = df_clean[col].replace({'nan': '', 'None': '', 'none': ''})
        df_clean[col] = df_clean[col].str.strip()
        # Set empty strings to NaN
        df_clean.loc[df_clean[col].str.len() < 2, col] = np.nan

    print("Step 6: Cleaning categorical data...")
    # Clean school names
    if 'which_school_do_you_attend' in df_clean.columns:
        df_clean['which_school_do_you_attend'] = df_clean['which_school_do_you_attend'].astype(str)
        # Standardize school names
        school_cleanup = {
            'worthing high': 'Worthing High School',
            'worthing high school': 'Worthing High School',
            'patcham high': 'Patcham High School',
            'patcham high school': 'Patcham High School',
            'seaford head': 'Seaford Head School',
            'seaford head school': 'Seaford Head School',
            'nan': 'Unknown'
        }

        for old, new in school_cleanup.items():
            df_clean['which_school_do_you_attend'] = df_clean['which_school_do_you_attend'].str.replace(
                old, new, case=False, regex=False
            )

    print("Step 7: Final cleanup...")
    # Reset index and remove duplicates
    df_clean = df_clean.reset_index(drop=True)
    df_clean = df_clean.drop_duplicates()

    return df_clean

# Apply cleaning
print("STARTING DATA CLEANING PROCESS...")
df_clean = clean_dataframe(df)
print("DATA CLEANING COMPLETED!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Load the Excel file
file_path = "shared_data_read_only/Data/BreckFoundation/InternalData/YA_post-assembly_questionnaire.xlsx"
df = pd.read_excel(file_path, sheet_name="Previous Responses")

print(f"Dataset loaded with {len(df)} rows")

# Combine text from key columns
text_columns = [
    'What have you learnt that you didn\'t know before?',
    'What aspect of Breck\'s story has had the biggest impact on you so far?',
    'Please explain your answer further if you can',
    'Is there anything else you would like to add?'
]

# Use only columns that exist
available_columns = [col for col in text_columns if col in df.columns]
print(f"Using columns: {available_columns}")

# Combine all text
all_text = ""
for col in available_columns:
    # Get non-empty responses
    text_data = df[col].dropna().astype(str)
    # Remove empty strings and "nan"
    text_data = text_data[text_data.str.len() > 0]
    text_data = text_data[text_data != 'nan']
    all_text += ' '.join(text_data) + ' '

print(f"Combined text length: {len(all_text)} characters")

# Simple text cleaning
def clean_text(text):
    text = text.lower()
    # Keep letters, spaces, and basic punctuation
    text = re.sub(r'[^\w\s]', ' ', text)
    text = ' '.join(text.split())  # Remove extra spaces
    return text

cleaned_text = clean_text(all_text)
print(f"After cleaning: {len(cleaned_text)} characters")

# Get word frequencies
words = cleaned_text.split()
print(f"Total words: {len(words)}")
print(f"Unique words: {len(set(words))}")

# Show all words before filtering
print("\nAll words found:")
word_freq = Counter(words)
print(f"Top 50 raw words: {word_freq.most_common(50)}")

# Minimal stop words - only the most common
minimal_stop_words = {'the', 'and', 'to', 'a', 'of', 'in', 'i', 'it', 'is', 'was', 'for'}

# Filter words
filtered_words = {}
for word, count in word_freq.items():
    if (word not in minimal_stop_words and
        len(word) > 2 and
        count >= 2):  # Only include words that appear at least twice
        filtered_words[word] = count

print(f"\nAfter filtering: {len(filtered_words)} words")
print(f"Filtered words: {filtered_words}")

# If we still have no words, use a less aggressive approach
if len(filtered_words) == 0:
    print("No words after filtering, using less aggressive approach...")
    filtered_words = {word: count for word, count in word_freq.items()
                     if len(word) > 2 and count >= 2}
    print(f"Now we have {len(filtered_words)} words")

# Create visualization
if len(filtered_words) > 0:
    # Get top words
    top_words = sorted(filtered_words.items(), key=lambda x: x[1], reverse=True)[:30]

    # Create horizontal bar chart
    plt.figure(figsize=(12, 10))
    words, counts = zip(*top_words)

    bars = plt.barh(words, counts, color='skyblue')
    plt.xlabel('Frequency')
    plt.title('Most Common Words in Assembly Feedback', fontsize=14, pad=20)
    plt.gca().invert_yaxis()  # Highest frequency at top

    # Add value labels on bars
    for bar, count in zip(bars, counts):
        width = bar.get_width()
        plt.text(width + 0.1, bar.get_y() + bar.get_height()/2,
                f'{count}', ha='left', va='center', fontsize=10)

    plt.tight_layout()
    plt.show()

    # Print summary
    print(f"\n=== TOP {len(top_words)} WORDS ===")
    for i, (word, count) in enumerate(top_words, 1):
        print(f"{i:2d}. {word:15} : {count}")

    # Analyze themes
    print("\n=== KEY THEMES ===")
    theme_keywords = {
        'Grooming': ['grooming', 'groomed', 'online', 'safety'],
        'Breck Story': ['breck', 'story', 'family', 'quickly', 'escalated'],
        'Learning': ['learnt', 'learned', 'understand', 'know', 'aware'],
        'Peer Impact': ['peer', 'child', 'listened', 'carefully', 'relatable']
    }

    for theme, keywords in theme_keywords.items():
        theme_count = sum(filtered_words.get(keyword, 0) for keyword in keywords)
        if theme_count > 0:
            print(f"{theme}: {theme_count} mentions")

else:
    print("No suitable words found for visualization.")
    print("Raw word frequency sample:")
    print(word_freq.most_common(50))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
from collections import Counter
import numpy as np

# Load your Excel file
file_path = "shared_data_read_only/Data/BreckFoundation/InternalData/YA_post-assembly_questionnaire.xlsx"
df = pd.read_excel(file_path, sheet_name="Previous Responses")

print(f"Dataset loaded: {len(df)} rows, {len(df.columns)} columns")

# Combine text from all columns that likely contain responses
all_text = ""

# Add text from each column
for col in df.columns:
    try:
        # Get non-empty values and convert to string
        text_data = df[col].dropna().astype(str)
        # Remove empty strings and 'nan'
        text_data = text_data[text_data.str.len() > 0]
        text_data = text_data[text_data != 'nan']

        if len(text_data) > 0:
            column_text = ' '.join(text_data)
            all_text += column_text + ' '
            print(f"Added text from column: {col[:50]}... ({len(text_data)} responses)")
    except Exception as e:
        continue

print(f"\nTotal text collected: {len(all_text)} characters")

if len(all_text) == 0:
    print("No text found. Trying alternative approach...")
    # Alternative: combine ALL cell values
    for col in df.columns:
        for cell in df[col].values:
            if pd.notna(cell):
                all_text += str(cell) + ' '
    print(f"Alternative text collection: {len(all_text)} characters")

# Clean the text
def clean_text(text):
    text = text.lower()
    # Remove numbers and special characters, keep letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    # Remove extra whitespace
    text = ' '.join(text.split())
    return text

cleaned_text = clean_text(all_text)
print(f"Cleaned text: {len(cleaned_text)} characters")

if len(cleaned_text) > 0:
    # Get word frequencies
    words = cleaned_text.split()
    word_freq = Counter(words)

    print(f"\nTotal words: {len(words)}")
    print(f"Unique words: {len(word_freq)}")

    # Show top words
    print("\nTop 20 words before filtering:")
    for word, count in word_freq.most_common(20):
        print(f"  {word}: {count}")

    # Filter out common stop words
    stop_words = {
        'the', 'and', 'to', 'a', 'of', 'in', 'i', 'it', 'is', 'was', 'for', 'with',
        'on', 'he', 'his', 'her', 'she', 'they', 'them', 'we', 'you', 'me', 'my',
        'your', 'our', 'their', 'this', 'that', 'these', 'those', 'be', 'been',
        'have', 'has', 'had', 'do', 'does', 'did', 'but', 'or', 'as', 'if', 'so',
        'than', 'then', 'just', 'like', 'more', 'most', 'very', 'what', 'when',
        'where', 'who', 'which', 'why', 'how', 'all', 'any', 'both', 'each'
    }

    # Filter words
    filtered_words = {}
    for word, count in word_freq.items():
        if (word not in stop_words and
            len(word) > 2 and
            count >= 2):  # Only include words that appear at least twice
            filtered_words[word] = count

    print(f"\nWords after filtering: {len(filtered_words)}")

    if len(filtered_words) > 0:
        # Create word cloud
        wordcloud = WordCloud(
            width=1200,
            height=800,
            background_color='white',
            colormap='plasma',
            max_words=100,
            relative_scaling=0.5,
            stopwords=set(),  # We already filtered stopwords
            collocations=False
        ).generate_from_frequencies(filtered_words)

        # Display the word cloud
        plt.figure(figsize=(15, 10))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud: Breck Foundation Assembly Feedback',
                 fontsize=20, pad=20, weight='bold')
        plt.tight_layout()
        plt.show()

        # Save the word cloud
        wordcloud.to_file("breck_assembly_wordcloud.png")
        print("Word cloud saved as 'breck_assembly_wordcloud.png'")

        # Show top words in the cloud
        print("\nTop words in the word cloud:")
        top_words = sorted(filtered_words.items(), key=lambda x: x[1], reverse=True)[:15]
        for word, count in top_words:
            print(f"  {word}: {count}")

    else:
        print("No words remaining after filtering. Creating word cloud with all words...")
        # Create with all words (minimal filtering)
        minimal_words = {word: count for word, count in word_freq.items()
                        if len(word) > 2 and count >= 1}

        if len(minimal_words) > 0:
            wordcloud = WordCloud(
                width=1200,
                height=800,
                background_color='white',
                colormap='viridis',
                max_words=80
            ).generate_from_frequencies(minimal_words)

            plt.figure(figsize=(15, 10))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            plt.title('Word Cloud: Assembly Feedback (All Words)',
                     fontsize=20, pad=20, weight='bold')
            plt.tight_layout()
            plt.show()

            wordcloud.to_file("breck_assembly_wordcloud_all_words.png")
            print("Word cloud saved as 'breck_assembly_wordcloud_all_words.png'")

else:
    print("No text data available to create word cloud")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
import numpy as np
from textblob import TextBlob
import warnings
warnings.filterwarnings('ignore')

# Load the data
file_path = "shared_data_read_only/Data/BreckFoundation/InternalData/YA_post-assembly_questionnaire.xlsx"
df = pd.read_excel(file_path, sheet_name="Previous Responses")

print("=== SEMANTIC ANALYSIS OF BRECK FOUNDATION ASSEMBLY FEEDBACK ===\n")

# 1. SENTIMENT ANALYSIS
print("1. SENTIMENT ANALYSIS")
print("=" * 50)

def analyze_sentiment(text):
    """Analyze sentiment of text using TextBlob"""
    if pd.isna(text) or str(text).strip() == '':
        return None
    blob = TextBlob(str(text))
    return blob.sentiment.polarity  # -1 (negative) to +1 (positive)

# Combine key text columns for sentiment analysis
text_columns_for_sentiment = []
for col in df.columns:
    if any(keyword in str(col).lower() for keyword in ['learn', 'understand', 'impact', 'explain', 'add']):
        text_columns_for_sentiment.append(col)

all_responses = []
sentiment_scores = []

for col in text_columns_for_sentiment[:3]:  # Use first 3 relevant columns
    for response in df[col].dropna():
        response_text = str(response).strip()
        if len(response_text) > 10:  # Only meaningful responses
            sentiment = analyze_sentiment(response_text)
            if sentiment is not None:
                sentiment_scores.append(sentiment)
                all_responses.append(response_text)

if sentiment_scores:
    avg_sentiment = np.mean(sentiment_scores)
    print(f"Average Sentiment Score: {avg_sentiment:.3f}")

    # Categorize sentiment
    positive = len([s for s in sentiment_scores if s > 0.1])
    neutral = len([s for s in sentiment_scores if -0.1 <= s <= 0.1])
    negative = len([s for s in sentiment_scores if s < -0.1])

    total = len(sentiment_scores)
    print(f"Positive responses: {positive}/{total} ({positive/total*100:.1f}%)")
    print(f"Neutral responses: {neutral}/{total} ({neutral/total*100:.1f}%)")
    print(f"Negative responses: {negative}/{total} ({negative/total*100:.1f}%)")

    # Plot sentiment distribution
    plt.figure(figsize=(10, 6))
    plt.hist(sentiment_scores, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    plt.axvline(avg_sentiment, color='red', linestyle='--', linewidth=2, label=f'Average: {avg_sentiment:.3f}')
    plt.xlabel('Sentiment Polarity (-1 = Negative, +1 = Positive)')
    plt.ylabel('Number of Responses')
    plt.title('Distribution of Sentiment in Assembly Feedback')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# 2. THEMATIC ANALYSIS
print("\n2. THEMATIC ANALYSIS")
print("=" * 50)

# Combine all text for thematic analysis
all_text = " ".join([str(x) for x in df.values.flatten() if pd.notna(x)])

# Define themes and their keywords
themes = {
    'Online Safety & Grooming': [
        'grooming', 'groomed', 'online', 'safety', 'risks', 'danger', 'safe',
        'strangers', 'trust', 'manipulation', 'vulnerable'
    ],
    'Breck Story Impact': [
        'breck', 'story', 'impact', 'family', 'affected', 'shocking', 'tragic',
        'murdered', 'killed', 'death'
    ],
    'Learning & Awareness': [
        'learnt', 'learned', 'understand', 'knowledge', 'aware', 'awareness',
        'realise', 'realize', 'know', 'knew'
    ],
    'Peer Delivery Effectiveness': [
        'peer', 'child', 'age', 'young', 'people', 'students', 'ambassadors',
        'listened', 'carefully', 'understood', 'relatable', 'relate',
        'delivery', 'explained', 'presented'
    ],
    'Emotional Response': [
        'sad', 'shocked', 'emotional', 'processing', 'thinking', 'feel',
        'felt', 'impactful', 'powerful', 'moving'
    ],
    'Behavioral Change': [
        'careful', 'cautious', 'share', 'tell', 'warn', 'speak', 'talk',
        'change', 'different', 'careful'
    ]
}

# Count theme occurrences
theme_counts = {}
for theme, keywords in themes.items():
    count = 0
    for keyword in keywords:
        # Count occurrences (case insensitive)
        count += len(re.findall(r'\b' + re.escape(keyword) + r'\b', all_text.lower()))
    theme_counts[theme] = count

# Display theme analysis
print("Theme Frequency Analysis:")
for theme, count in sorted(theme_counts.items(), key=lambda x: x[1], reverse=True):
    if count > 0:
        print(f"  {theme}: {count} mentions")

# Plot theme analysis
plt.figure(figsize=(12, 8))
themes_sorted = dict(sorted(theme_counts.items(), key=lambda x: x[1], reverse=True))
bars = plt.barh(list(themes_sorted.keys()), list(themes_sorted.values()), color='lightcoral')
plt.xlabel('Number of Mentions')
plt.title('Thematic Analysis of Assembly Feedback')
plt.gca().invert_yaxis()

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height()/2, f'{int(width)}',
             ha='left', va='center')

plt.tight_layout()
plt.show()

# 3. KEY PHRASE EXTRACTION
print("\n3. KEY PHRASE EXTRACTION")
print("=" * 50)

def extract_key_phrases(text, min_occurrence=3):
    """Extract common phrases and their frequencies"""
    # Simple n-gram extraction (2-word phrases)
    words = text.lower().split()
    phrases = []

    for i in range(len(words) - 1):
        phrase = f"{words[i]} {words[i + 1]}"
        # Only keep meaningful phrases (both words > 2 chars, not stop words)
        if (len(words[i]) > 2 and len(words[i + 1]) > 2 and
            words[i] not in ['the', 'and', 'for', 'with', 'this'] and
            words[i + 1] not in ['the', 'and', 'for', 'with', 'this']):
            phrases.append(phrase)

    phrase_counts = Counter(phrases)
    return {phrase: count for phrase, count in phrase_counts.items() if count >= min_occurrence}

key_phrases = extract_key_phrases(all_text)
if key_phrases:
    print("Most Common 2-Word Phrases:")
    for phrase, count in sorted(key_phrases.items(), key=lambda x: x[1], reverse=True)[:15]:
        print(f"  '{phrase}': {count} times")
else:
    print("No significant phrases found with minimum occurrence threshold")

# 4. IMPACT ASSESSMENT ANALYSIS
print("\n4. IMPACT ASSESSMENT")
print("=" * 50)

# Look for impact-related statements
impact_patterns = {
    'Changed Behavior': r'(more careful|will be careful|be careful|tell someone|share with|warn|speak up)',
    'Increased Awareness': r'(didn\'t know|never knew|learned that|realised|aware now|understand now)',
    'Emotional Impact': r'(shocked|sad|emotional|affected me|made me think|impact on me)',
    'Peer Learning': r'(peer|same age|young people|students|ambassadors|relatable)',
    'Knowledge Gain': r'(learnt about|learned about|understand grooming|know about|aware of)'
}

impact_counts = {}
for pattern_name, pattern in impact_patterns.items():
    matches = re.findall(pattern, all_text.lower())
    impact_counts[pattern_name] = len(matches)

print("Impact Assessment:")
for impact_type, count in sorted(impact_counts.items(), key=lambda x: x[1], reverse=True):
    if count > 0:
        print(f"  {impact_type}: {count} mentions")

# 5. COMPARATIVE ANALYSIS: PEER vs ADULT DELIVERY
print("\n5. DELIVERY METHOD EFFECTIVENESS")
print("=" * 50)

# Extract delivery preference patterns
delivery_patterns = {
    'Preferred Peer Delivery': r'(listened more carefully|understood better from peer|better from peer|relatable|same age|young people)',
    'No Difference': r'(no difference|same either way|doesn\'t matter|same story|same words)',
    'Preferred Adult Delivery': r'(adult more|adult better|adult knowledgeable|understand from adult)'
}

delivery_analysis = {}
for pattern_name, pattern in delivery_patterns.items():
    matches = re.findall(pattern, all_text.lower())
    delivery_analysis[pattern_name] = len(matches)

print("Delivery Method Preferences:")
total_delivery_mentions = sum(delivery_analysis.values())
for pref, count in delivery_analysis.items():
    if total_delivery_mentions > 0:
        percentage = (count / total_delivery_mentions) * 100
        print(f"  {pref}: {count} mentions ({percentage:.1f}%)")

# Plot delivery preferences
if total_delivery_mentions > 0:
    plt.figure(figsize=(10, 6))
    preferences = list(delivery_analysis.keys())
    counts = list(delivery_analysis.values())

    colors = ['lightgreen', 'lightblue', 'lightcoral']
    bars = plt.bar(preferences, counts, color=colors, alpha=0.8)

    plt.ylabel('Number of Mentions')
    plt.title('Delivery Method Preferences\n(Peer vs Adult-led Assembly)')
    plt.xticks(rotation=15, ha='right')

    # Add value labels on bars
    for bar, count in zip(bars, counts):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{count}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

# 6. SUMMARY INSIGHTS
print("\n6. KEY INSIGHTS SUMMARY")
print("=" * 50)

# Generate insights based on analysis
insights = []

if sentiment_scores and avg_sentiment > 0:
    insights.append(" Overall positive reception to the assembly")
elif sentiment_scores and avg_sentiment < 0:
    insights.append(" Mixed to slightly negative emotional response")
else:
    insights.append(" Generally neutral to positive engagement")

# Theme-based insights
top_themes = sorted(theme_counts.items(), key=lambda x: x[1], reverse=True)[:3]
if top_themes:
    insight_msg = " Top themes: " + ", ".join([theme for theme, count in top_themes if count > 0])
    insights.append(insight_msg)

# Delivery method insight
if delivery_analysis.get('Preferred Peer Delivery', 0) > delivery_analysis.get('Preferred Adult Delivery', 0):
    insights.append(" Strong preference for peer-led delivery over adult-led")
elif delivery_analysis.get('Preferred Adult Delivery', 0) > delivery_analysis.get('Preferred Peer Delivery', 0):
    insights.append(" Preference for adult-led delivery")
else:
    insights.append(" Mixed opinions on delivery method")

# Impact insight
if impact_counts.get('Changed Behavior', 0) > 5:
    insights.append(" Assembly successfully influenced behavioral intentions")
if impact_counts.get('Increased Awareness', 0) > 5:
    insights.append(" Significant knowledge gain reported by participants")

# Print insights
for insight in insights:
    print(insight)

print(f"\n=== ANALYSIS COMPLETE ===")
print(f"Total responses analyzed: {len(df)}")
print(f"Text corpus size: {len(all_text)} characters")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import re

# Load the data
file_path = "shared_data_read_only/Data/BreckFoundation/InternalData/YA_post-assembly_questionnaire.xlsx"
df = pd.read_excel(file_path, sheet_name="Previous Responses")

print("=== COMPARATIVE ANALYSIS OF SCHOOLS ===")
print("=" * 50)

# 1. EXTRACT AND CLEAN SCHOOL DATA
print("1. SCHOOL DISTRIBUTION")
print("=" * 50)

# Find the school column (looking for variations)
school_column = None
for col in df.columns:
    if 'school' in str(col).lower():
        school_column = col
        break

if school_column is None:
    # If no school column found, look for school names in all columns
    print("No dedicated school column found. Searching all columns for school names...")
    all_text = " ".join([str(x) for x in df.values.flatten() if pd.notna(x)])
else:
    print(f"Using school column: '{school_column}'")

# Extract schools from the data
schools_data = []
if school_column and school_column in df.columns:
    schools_data = df[school_column].dropna().astype(str).tolist()
else:
    # Search through all columns for school mentions
    for col in df.columns:
        for value in df[col].dropna():
            text = str(value).lower()
            if any(school in text for school in ['worthing', 'patcham', 'seaford']):
                schools_data.append(text)

# Clean and categorize schools
school_categories = {
    'Worthing High': [],
    'Patcham High': [],
    'Seaford Head': []
}

for school_text in schools_data:
    school_text_lower = school_text.lower()
    if 'worthing' in school_text_lower:
        school_categories['Worthing High'].append(school_text)
    elif 'patcham' in school_text_lower:
        school_categories['Patcham High'].append(school_text)
    elif 'seaford' in school_text_lower:
        school_categories['Seaford Head'].append(school_text)

# Count students per school
school_counts = {school: len(entries) for school, entries in school_categories.items()}
total_students = sum(school_counts.values())

print(f"Total students with school identification: {total_students}")
for school, count in school_counts.items():
    percentage = (count / total_students) * 100 if total_students > 0 else 0
    print(f"  {school}: {count} students ({percentage:.1f}%)")

# 2. COMPARATIVE ANALYSIS BY SCHOOL
print("\n2. COMPARATIVE ANALYSIS BY SCHOOL")
print("=" * 50)

# Create a function to extract responses by school
def get_school_responses(school_name):
    """Get all responses from a specific school"""
    school_responses = []
    for idx, row in df.iterrows():
        row_text = " ".join([str(x) for x in row.values if pd.notna(x)])
        if school_name.lower() in row_text.lower():
            school_responses.append(row_text)
    return school_responses

# Get responses for each school
worthing_responses = get_school_responses('worthing')
patcham_responses = get_school_responses('patcham')
seaford_responses = get_school_responses('seaford')

print(f"Worthing High: {len(worthing_responses)} responses")
print(f"Patcham High: {len(patcham_responses)} responses")
print(f"Seaford Head: {len(seaford_responses)} responses")

# 3. KEY METRICS COMPARISON
print("\n3. KEY METRICS COMPARISON")
print("=" * 50)

def analyze_school_metrics(responses, school_name):
    """Analyze key metrics for each school"""
    if not responses:
        return {}

    all_text = " ".join(responses).lower()

    # Calculate metrics
    metrics = {
        'response_count': len(responses),
        'total_words': len(all_text.split()),
        'avg_words_per_response': len(all_text.split()) / len(responses) if responses else 0
    }

    # Key theme frequencies
    themes = {
        'grooming_mentions': len(re.findall(r'grooming|groomed', all_text)),
        'online_safety': len(re.findall(r'online|safety|safe|danger', all_text)),
        'peer_delivery': len(re.findall(r'peer|child|age|young|student', all_text)),
        'emotional_impact': len(re.findall(r'shock|sad|emotional|impact|affect', all_text)),
        'learning_gain': len(re.findall(r'learn|understand|know|aware', all_text)),
        'breck_family': len(re.findall(r'family|mother|parent', all_text))
    }

    # Normalize by number of responses
    for theme in themes:
        themes[theme] = themes[theme] / len(responses) if responses else 0

    metrics['themes'] = themes
    return metrics

# Analyze each school
worthing_metrics = analyze_school_metrics(worthing_responses, 'Worthing High')
patcham_metrics = analyze_school_metrics(patcham_responses, 'Patcham High')
seaford_metrics = analyze_school_metrics(seaford_responses, 'Seaford Head')

# Display comparative metrics
print("\nTheme Frequency (per response):")
themes_list = ['grooming_mentions', 'online_safety', 'peer_delivery', 'emotional_impact', 'learning_gain', 'breck_family']
theme_labels = ['Grooming', 'Online Safety', 'Peer Delivery', 'Emotional Impact', 'Learning', 'Breck Family']

for i, theme in enumerate(themes_list):
    w_val = worthing_metrics.get('themes', {}).get(theme, 0)
    p_val = patcham_metrics.get('themes', {}).get(theme, 0)
    s_val = seaford_metrics.get('themes', {}).get(theme, 0)

    print(f"\n{theme_labels[i]}:")
    print(f"  Worthing: {w_val:.2f}")
    print(f"  Patcham:  {p_val:.2f}")
    print(f"  Seaford:  {s_val:.2f}")

# 4. VISUAL COMPARISON
print("\n4. VISUAL COMPARISON")
print("=" * 50)

# Create comparative visualizations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Response Count by School
schools = ['Worthing High', 'Patcham High', 'Seaford Head']
counts = [worthing_metrics.get('response_count', 0),
          patcham_metrics.get('response_count', 0),
          seaford_metrics.get('response_count', 0)]

bars1 = ax1.bar(schools, counts, color=['#1f77b4', '#ff7f0e', '#2ca02c'], alpha=0.8)
ax1.set_title('Number of Responses by School', fontsize=14, weight='bold')
ax1.set_ylabel('Number of Responses')
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom')

# Plot 2: Theme Comparison
theme_data = []
for theme in themes_list:
    theme_data.append([
        worthing_metrics.get('themes', {}).get(theme, 0),
        patcham_metrics.get('themes', {}).get(theme, 0),
        seaford_metrics.get('themes', {}).get(theme, 0)
    ])

x = np.arange(len(theme_labels))
width = 0.25

for i, school_data in enumerate(np.array(theme_data).T):
    ax2.bar(x + i*width, school_data, width, label=schools[i], alpha=0.8)

ax2.set_title('Theme Frequency Comparison (per response)', fontsize=14, weight='bold')
ax2.set_ylabel('Mentions per Response')
ax2.set_xticks(x + width)
ax2.set_xticklabels(theme_labels, rotation=45, ha='right')
ax2.legend()

# Plot 3: Word Cloud of Unique Words by School
def get_top_words(responses, top_n=10):
    """Get top unique words for each school"""
    if not responses:
        return []
    all_text = " ".join(responses).lower()
    words = re.findall(r'\b[a-z]{4,}\b', all_text)
    # Remove common words
    stop_words = {'that', 'with', 'this', 'from', 'they', 'have', 'more', 'were', 'been', 'like'}
    filtered_words = [w for w in words if w not in stop_words]
    word_freq = Counter(filtered_words)
    return word_freq.most_common(top_n)

worthing_words = get_top_words(worthing_responses)
patcham_words = get_top_words(patcham_responses)
seaford_words = get_top_words(seaford_responses)

# Display top words in table format
word_data = []
max_words = max(len(worthing_words), len(patcham_words), len(seaford_words))
for i in range(max_words):
    row = []
    row.append(worthing_words[i][0] if i < len(worthing_words) else "")
    row.append(worthing_words[i][1] if i < len(worthing_words) else 0)
    row.append(patcham_words[i][0] if i < len(patcham_words) else "")
    row.append(patcham_words[i][1] if i < len(patcham_words) else 0)
    row.append(seaford_words[i][0] if i < len(seaford_words) else "")
    row.append(seaford_words[i][1] if i < len(seaford_words) else 0)
    word_data.append(row)

# Create a table for top words
ax3.axis('tight')
ax3.axis('off')
table = ax3.table(cellText=word_data,
                 colLabels=['Worthing Word', 'Count', 'Patcham Word', 'Count', 'Seaford Word', 'Count'],
                 loc='center',
                 cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1, 1.5)
ax3.set_title('Top Unique Words by School', fontsize=14, weight='bold')

# Plot 4: Engagement Metrics
engagement_metrics = ['Avg Words/Response', 'Grooming Focus', 'Learning Focus']
worthing_engagement = [
    worthing_metrics.get('avg_words_per_response', 0),
    worthing_metrics.get('themes', {}).get('grooming_mentions', 0),
    worthing_metrics.get('themes', {}).get('learning_gain', 0)
]
patcham_engagement = [
    patcham_metrics.get('avg_words_per_response', 0),
    patcham_metrics.get('themes', {}).get('grooming_mentions', 0),
    patcham_metrics.get('themes', {}).get('learning_gain', 0)
]
seaford_engagement = [
    seaford_metrics.get('avg_words_per_response', 0),
    seaford_metrics.get('themes', {}).get('grooming_mentions', 0),
    seaford_metrics.get('themes', {}).get('learning_gain', 0)
]

x_eng = np.arange(len(engagement_metrics))
width_eng = 0.25

bars4a = ax4.bar(x_eng - width_eng, worthing_engagement, width_eng, label='Worthing', alpha=0.8)
bars4b = ax4.bar(x_eng, patcham_engagement, width_eng, label='Patcham', alpha=0.8)
bars4c = ax4.bar(x_eng + width_eng, seaford_engagement, width_eng, label='Seaford', alpha=0.8)

ax4.set_title('Engagement Metrics Comparison', fontsize=14, weight='bold')
ax4.set_ylabel('Score')
ax4.set_xticks(x_eng)
ax4.set_xticklabels(engagement_metrics, rotation=45, ha='right')
ax4.legend()

plt.tight_layout()
plt.show()

# 5. KEY FINDINGS BY SCHOOL
print("\n5. KEY FINDINGS BY SCHOOL")
print("=" * 50)

def generate_school_insights(metrics, school_name):
    """Generate insights for each school"""
    if metrics.get('response_count', 0) == 0:
        return f"No sufficient data for {school_name}"

    insights = []
    themes = metrics.get('themes', {})

    # Engagement level
    avg_words = metrics.get('avg_words_per_response', 0)
    if avg_words > 30:
        insights.append("High engagement level (detailed responses)")
    elif avg_words > 15:
        insights.append("Moderate engagement level")
    else:
        insights.append("Brief response style")

    # Top themes
    if themes.get('grooming_mentions', 0) > 0.5:
        insights.append("Strong focus on grooming aspects")
    if themes.get('emotional_impact', 0) > 0.3:
        insights.append("Notable emotional impact reported")
    if themes.get('peer_delivery', 0) > 0.4:
        insights.append("High awareness of peer delivery method")
    if themes.get('learning_gain', 0) > 0.5:
        insights.append("Significant learning outcomes")

    return insights

print("\nWORTHING HIGH SCHOOL:")
worthing_insights = generate_school_insights(worthing_metrics, 'Worthing High')
for insight in worthing_insights:
    print(f"   {insight}")

print("\nPATCHAM HIGH SCHOOL:")
patcham_insights = generate_school_insights(patcham_metrics, 'Patcham High')
for insight in patcham_insights:
    print(f"   {insight}")

print("\nSEAFORD HEAD SCHOOL:")
seaford_insights = generate_school_insights(seaford_metrics, 'Seaford Head')
for insight in seaford_insights:
    print(f"   {insight}")

# 6. RECOMMENDATIONS
print("\n6. SCHOOL-SPECIFIC RECOMMENDATIONS")
print("=" * 50)

print("\nBased on the analysis, here are tailored recommendations:")

# Generate recommendations based on patterns
if worthing_metrics.get('response_count', 0) > 0:
    print(f"\nFor WORTHING HIGH:")
    if worthing_metrics.get('themes', {}).get('peer_delivery', 0) > 0.4:
        print("   Continue peer-led delivery approach - it's resonating well")
    if worthing_metrics.get('themes', {}).get('learning_gain', 0) < 0.3:
        print("   Consider reinforcing key learning points in follow-up sessions")

if patcham_metrics.get('response_count', 0) > 0:
    print(f"\nFor PATCHAM HIGH:")
    if patcham_metrics.get('themes', {}).get('emotional_impact', 0) > 0.4:
        print("   Provide additional support resources for emotionally affected students")
    print("   The balanced response pattern suggests good overall reception")

if seaford_metrics.get('response_count', 0) > 0:
    print(f"\nFor SEAFORD HEAD:")
    if seaford_metrics.get('themes', {}).get('grooming_mentions', 0) < 0.3:
        print("   Emphasize grooming awareness in future sessions")
    print("   Consider gathering more detailed feedback to understand specific impacts")

print(f"\n=== ANALYSIS COMPLETE ===")
print(f"Total responses analyzed: {len(df)}")
print(f"Schools with identifiable data: {total_students}")
```
